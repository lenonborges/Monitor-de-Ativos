```python
# Passando a lista para DataFrame
df = pd.DataFrame(data=lista)
df
```

```python
# Loop p pegar preço e variação de cada ação usando o yfinance
aj = []
lm12=[]
lm6=[]
lm3=[]
lm1=[]
lw2=[]
for i in df[0]:
  msft = yf.Ticker(i)
  data = msft.history(period='12mo')
  # pegando o volume médio do último mês
  j = data[['Volume'][-1:22]].mean() 
  j = j[0]
  # condição mínima para o número de ações negociadas da ação
  if j > 20000: 
    aj.append(i)
    try:
      m12 = (((data['Close'][-1]/data['Close'][0])-1)*100)
      lm12.append(m12)
    except:
      m12 = 'N/A'
      lm12.append(m12)
      pass
    try:
      m6 = (((data['Close'][-1]/data['Close'][-126])-1)*100) #126 por ser metade de 252 e me trazer o retorno em 6 meses
      lm6.append(m6)
    except:
      m6 = 'N/A'
      lm6.append(m6)
      pass
    try:
      m3 = (((data['Close'][-1]/data['Close'][-63])-1)*100)
      lm3.append(m3)
    except:
      m3 = 'N/A'
      lm3.append(m3)
      pass
    try:
      m1 = (((data['Close'][-1]/data['Close'][-21])-1)*100)
      lm1.append(m1)
    except:
      m1 = 'N/A'
      lm1.append(m1)
      pass
    try:
      w2 = (((data['Close'][-1]/data['Close'][-10])-1)*100)
      lw2.append(w2)
    except:
      w2 = 'N/A'
      lw2.append(w2)
      pass
  else:
    pass
```
